```python
import requests
import json

cotacoes = requests.get('https://economia.awesomeapi.com.br/json/all')
cotacoes_dic = cotacoes.json()
print(cotacoes_dic)
```

### Pegar a Cotação Atual de Todas as Moedas

```python
for chave, valor in cotacoes_dic.items():
    print(f'A cotação do {chave} é de ', end = '')
    for parametro, cotacao in valor.items():
        if parametro == 'bid':
            print(cotacao)
```

### A última cotação do Dólar, do Euro e do BitCoin?

```python
print('A última cotação do Dólar foi de {}'.format(cotacoes_dic['USD']['bid']))
print('A última cotação do Euro foi de {}'.format(cotacoes_dic['EUR']['bid']))
print('A última cotação do Bitcoin foi de {}'.format(cotacoes_dic['BTC']['bid']))
```

### A cotação dos últimos 30 dias do dólar 

```python
cotacoes_30dolar = requests.get('https://economia.awesomeapi.com.br/json/daily/USD-BRL/30')
cotacoes_dolar = cotacoes_30dolar.json()

resultado_dolar30 = [codigo['bid'] for codigo in cotacoes_dolar]
print(resultado_dolar30)
```

### As cotações do BitCoin de Jan/20 a Out/20

```python
cotacoes_btc_mes = requests.get('https://economia.awesomeapi.com.br/json/daily/BTC-BRL/200?start_date=20200101&end_date=20201031')
cotacoes_btc = cotacoes_btc_mes.json()

resultado_btc = [float(btc['bid']) for btc in cotacoes_btc]
resultado_btc.reverse()
print(resultado_btc)
```

### Gráfico com as cotações do BitCoin

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(17,7))
plt.plot(resultado_btc)
plt.show()
```
